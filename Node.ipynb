# Graph problem

Graph problem is a way to  simulate a real life problem into the digital world. For example, let's say you want to go from your house to the local shop,this is a real life “problem” we can translate that into a graph problem by converting the “house” into a start node and the “shop” to a goal node or state.in formal more academic definition graph problem is a mathematical construct that is used for modeling a real-world problem by dividing the problem into a set of connected nodes.The following graph will give you a simple idea on what do we mean by Nodes. we will talk in depth on links between them in the following section


As you can see , this is what we call a **directed graph**. Each **node** that is displayed can represent a state, maybe node 1 can be the **start state ”house”** and node 2 can be a “gas station”. Now if you pay attention you will notice a line between each node. This is called **edge** and it's a way of linking the nodes with each other. An edge can have a **cost**, it can be time , distance or number of steps. it's a variable that will change on each problem in the directed graph and it's displayed on the edges. The edges have **directions**, there is an edge from node 1 to node 2 but I can't go from node 2 to node 1 because the arrow is pointing to node 2 from node 1 same goes for node 4. We can also have a **Start state** and a **goal state**. A start state is where the problem will begin your initial position in the problem and the goal state is the state or the node that you want to reach.

# A*
A* (A star) algorithm is a search algorithm that finds the shortest path from the starting node to the goal node. It combines two algorithms: **greedy search algorithm** and **uniform cost search algorithm** to form one reliable algorithm.A* is considered an **informed search algorithm**. informed search algorithms are implemented using something called **“heuristic function”**. Heuristic function is like a prediction mechanism to predict the cost of an ongoing edge , so we don't have to reach unnecessary nodes.As shown in the upper figure we have a **heuristic table** which gives us the heuristic of each node, but as you can see the heuristic does not equal the actual cost which is normal because its a prediction it does have to be 100%. the heuristic of node B is 4 and the heuristic of node A  is 12 so the algorithm will chose node B and so on . 

- g(n) — this represents the exact cost of the path from the starting node to any node n.
- h(n) — this represents the heuristic estimated cost from node n to the goal node.
- f(n) — lowest cost in the neighboring node n ,f(n)=g(n)+h(n).
